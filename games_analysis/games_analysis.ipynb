# Анализ игр

Цель - выявить определяющие успешность игры закономерности, чтобы сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Определять потенциал будем на 2017 год.

Описание данных:


Данные до 2016 года из открытых источников. Данные за 2016 год могут быть неполными.

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Подготовка данных

### Столбцы

Приведем названия столбцов к нижнему регистру

````python
df.columns = df.columns.str.lower()
df.columns
````

### Пропуски и типы

**name**

````python
df.query('name.isna()')  # Cтроки с пропусками в названии (совпадают с пропусками в жанре)
````

````python
df = df.dropna(subset='name')  # Удалим строки с пропусками в названиях, тк они не представляют ценности
````

**year_of_release**

Корректно заменить пропуски в столбце не представляется возможным, поэтому удалим их, так как процент очень мал (менее 2%).
И переведем в целочисленный тип

````python
df = df.dropna(subset='year_of_release')  
df.year_of_release =  df.year_of_release.astype('uint16')
df.reset_index(drop=True);     # Cбросим индексы
````

**critic_score**

````python
df.critic_score.unique()
````

Корректно заменить пропуски никак нельзя, поэтому оставим как есть

**user_score**

Приведем столбец к числовому типу. Так как помимо чисел встречаются значения `tbd` (to be decided), то есть оценки не были поставлены, используем агрумент `coerce` для перевода их в NaN. Заменить пропуски также не представляется возможным.

````python
df.user_score = pd.to_numeric(df.user_score, errors='coerce')
````

**rating**

````python
df.rating.value_counts()
````

````python
df.query('rating == "AO"')
````

````python
df.rating = df.rating.fillna('unknown')
df.rating = df.rating.replace({'K-A':'E',
                              'RP':'unknown',
                              'AO':'M'})
````

- Объединили все неизвестные рейтинги (NaN) в группу "unknown", для удобства анализа.
- Согласно официальной документации, рейтинг "K-A" устаревший, его заменили на "E", мы тоже так сделали.
- RP означает рейтинг не присвоен. Заменили на "unknown".
- Единственный "AO" (Adults Only) присвоен игре GTA: San Andreas временно, после некоторых поправок в игре ей вернули рейтинг "M". Заменили аналогично.

### world_sales

В новом столбце `world_sales` посчитаем суммарное количество продаж по всем регионам.

````python
df['world_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
````

- Привели названия столбцов к нижнему регистру.

- В столбцe `year_of_release` заменили пропуски на отличающиеся от остальных значения (1900), чтобы в дальнейшем их использовать для анализа. Привели к целочисленному типу.
- В `critic_score` заменить пропуски корректно никак нельзя и удалить тоже(их много), поэтому оставляем.
- В столбце `user_score` заменили значения tbd на NaN. Привели к типу float. Зменить или удалить нельзя.
- В столбце `rating` заменили пропуски на unknown, как и RP(не присвоен), K-A(устаревший) заменили на E, AO(временный) на M.
- Создали новый столбец `world_sales` с суммарным количеством продаж по всем регионам.
- Дубликатов не обнаружено.

### Определим актуальный период времени

**Построим** график рапределения выпущенных игр по годам и список топ-10 по количеству

````python
df.groupby('year_of_release')['name'].count().plot(xlim=(1980, 2017), grid=True)
plt.title('Распределение выпущенных игр по годам')
plt.xlabel('год')
plt.ylabel('количество игр')
plt.show()

df.groupby('year_of_release')['name'].count().sort_values(ascending=False).head(10)
````

Наблюдаем основной рост с 2000 по 2008 год. Далее идет спад, но не опускается ниже уровня 2000 года. Пик 2008-2009 год

**Построим** графики распределения суммарного количетва продаж в мире по платформам и список топ-10 по сумме

````python
df.groupby('platform')['world_sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Распределение суммарного количетва продаж в мире по платформам')
plt.xlabel('платформа')
plt.ylabel('млн копий')
plt.show()

top_10_platform = df.groupby('platform')['world_sales'].sum().sort_values(ascending=False).head(10)
top_10_platform
````

Наблюдаем среднее "время жизни" платформ 5 лет. Пик популярности 1-2 года. Логично, ведь платформы устаревают, появляются новые более прогрессивные. Значит:
- Так как нас интересует потенциал на год грядущий, а динамика игр более специфична, будем рассматривать данные за последние 3 года, то есть 2014-2016 года.

- Платформа `PC` самая "долгоиграющая" на рынке. Наблюдается волнообразные рост и падения. Исключать потенциал персональных компьютеров нельзя, поскольку эта платформа более универсальна, чем игровые консоли. К тому же именно на ПК проводится большинство соревнований по киберспорту, а это большая площадка для рекламных кампаний. Поэтому также будем рассматривать как `актуальную` игровую платформу.

### Анализ платформ

**Создадим** новый датафрейм по срезу нашего старого, куда войдут данные за актуальный период и платформы с потенциалом

````python
actual_df = df.query('platform in ["PS4", "XOne", "PC"] and year_of_release >= 2014')
````

**Посмотрим** на распределение продаж по платформам

````python
ax = sns.boxplot(data=actual_df,
            x='platform',
            y='world_sales')

ax.set_title('Глобальные продажи')
ax.set(xlabel='', ylabel='млн копий')
plt.show()

ax = sns.boxplot(data=actual_df,
            x='platform',
            y='world_sales')

ax.set_title('Глобальные продажи')
ax.set(xlabel='', ylabel='млн копий', ylim=(0, 1))
plt.show()
````

Видим, что основная масса игр приходится на небольшое количество продаж. Медиана у PS4 и XOne примерно одинаковая, как и IQR, но вот аномально большие значения у PS4 гораздо выше. Это означает, что у PS4 есть немало гораздо более популярных игр. PC куда скромнее остальных, медиана на уровне 1-го квартиля, а 3-й квартиль на уровне медианы.

---

**Построим** графики и выясним корреляцию между оценками и глобальными продажами по каждой актуальной платформе

````python
# Цикл для создания графиков scatterplot и оценки корреляции
for pl in ['PS4', 'XOne', 'PC']:
    ax = (sns.scatterplot(data=actual_df.query('platform == @pl'),  # По оценкам критиков
                         x='critic_score',
                         y='world_sales'))
    ax.set_title('Взаимосвязь между оценкой критиков и продажами у ' + pl)
    ax.set(xlabel='оценка', ylabel='млн копий')
    plt.show()
    print('корреляция =', actual_df.query('platform == @pl')\
    .critic_score.corr(actual_df.query('platform == @pl').world_sales))

    ax = (sns.scatterplot(data=actual_df.query('platform == @pl'),  # По оценкам пользователей
                         x='user_score',
                         y='world_sales'))
    ax.set_title('Взаимосвязь между оценкой пользователей и продажами у ' + pl)
    ax.set(xlabel='оценка', ylabel='млн копий')
    plt.show()
    print('корреляция =', actual_df.query('platform == @pl')\
          .user_score.corr(actual_df.query('platform == @pl').world_sales))
````

<div style='border :solid SteelBlue 3px; padding:20px'>


- У `PS4` и `XOne` взаимосвязь между оценками пользователей и продажами отрицательная и очень слабая(-0.04,-0.07), можно сказать случайная. А вот с оценками критиков присутствует умеренная положительная(0.4,0.43).  
- У `PC` взаимосвязь между оценками пользователей и продажами очень слабая отрицательная(-0.07), а между оценками критиков и продажами слабая положительная(0.17). 
- Вероятно это связано с тем, что критики более основательно и объективно подходят к процессу. К тому же, как правило, они получают возможность оценить игровой процесс до выхода игры на рынок и некоторые пользователи ориентируются на эту оценку.

## Портрет пользователя по регионам

### Топ-5 платформ

**Построим** графики, отображающие топ-5 популярных платформ по регионам за актуальный период времени

````python
# Содаем цикл для построения графиков по регионам
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Топ-5 платформ по регионам')
for i in range(3):
    r = ['na_sales', 'eu_sales', 'jp_sales']
    n = ['Северная Америка', 'Европа', 'Япония']
    x = (df.query('year_of_release >= 2014')
        .groupby('platform')[r[i]].sum().reset_index()
        .sort_values(r[i], ascending=False, ignore_index=True)
        .head())
    x = x.append({'platform':'Other', r[i]:df.query('year_of_release >= 2014')[r[i]].sum() - x[r[i]].sum()},
                 ignore_index=True)  # Добавляем остальные в other
    axs[i].pie(x[r[i]], labels=x['platform'], autopct='%1.1f%%')
    axs[i].set_title(n[i])
plt.show()
````

- В Северной Америке распределение популярности платформ равномерное. По убыванию: `PS4`, `XOne`, `X360`, `PS3`, `3DS`.
- В Европе с отрывом лидирует `PS4`, далее равномерно по убыванию: `PS3`, `XOne`, `X360`, `3DS`.
- В Японии с отрывом лидирует `3DS`, далее равномерно по убыванию: `PS3`, `PSV`, `PS4`, `WiiU`.

### Топ-5 жанров

## Проверка гипотез

### Первая гипотеза

**Гипотеза**: средние пользовательские рейтинги платформ `XOne` и `PC` одинаковы.

```
H_0: Cредний рейтинг(user_score) "XOne" = средний рейтинг(user_score) "PC"
H_a: Cредний рейтинг(user_score) "XOne" ≠ средний рейтинг(user_score) "PC"
alpha = 0.05
```

**Применим** `t-test` для сравнения двух выборок по выручке, но сначала взгянем на `размер`, `дисперсию`, `среднее` и проверим выборочное распределение на нормальность тестом Шапиро-Уилка.

````python
display(actual_df[actual_df['platform'] == 'XOne']['user_score'].count(),  # Размер
actual_df[actual_df['platform'] == 'PC']['user_score'].count())
````

````python
display(np.var(actual_df[actual_df['platform'] == 'XOne']['user_score']),  # Дисперсия
np.var(actual_df[actual_df['platform'] == 'PC']['user_score']))
````

````python
display(actual_df[actual_df['platform'] == 'XOne']['user_score'].mean(),  # Среднее
actual_df[actual_df['platform'] == 'PC']['user_score'].mean())
````

````python
lst = []
for i in range(1000):
    lst.append(actual_df.query('platform == "XOne"')['user_score']\
                        .sample(160, replace=True).mean())
    
print(f'Шапиро p-value = {st.shapiro(lst).pvalue}')
sns.histplot(lst)
plt.title('XOne')
plt.show()
    
    
lst = []
for i in range(1000):
    lst.append(actual_df.query('platform == "PC"')['user_score']\
                        .sample(120, replace=True).mean())
        
print(f'Шапиро p-value = {st.shapiro(lst).pvalue}')
sns.histplot(lst)
plt.title('PC')
plt.show()
````

Наблюдаем разницу в размере и дисперсии. Нормальность выборочного распределения подтверждена.

---

````python
results = st.ttest_ind(actual_df[(actual_df['platform'] == 'XOne') & (actual_df['user_score'] >= 0)]['user_score'],
                       actual_df[(actual_df['platform'] == 'PC') & (actual_df['user_score'] >= 0)]['user_score'],
                       equal_var=False)  # Так как размер и дисперсии выборок отличаются

alpha = .05

print(results.pvalue) 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

P-value говорит нам о том, что разница между пользовательскими рейтингами у `XOne` и `PC` незначима. Следовательно можно предположить, что рейтинги одинаковы.

**Применим** `t-test` для сравнения двух выборок по выручке, но сначала `размер`, `дисперсия`, `среднее` и тест на нормальность Шапиро-Уилка.
