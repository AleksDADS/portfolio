# Исследование объявлений о продаже квартир

* Данные сервиса Яндекс Недвижимость — архив объявлений за несколько лет о продаже квартир в Санкт-Петербурге и соседних населённых пунктах.


* Цель — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости.

Описание данных  
О каждой квартире в базе содержится два типа данных: добавленные пользователем и картографические.

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

## Предобработка данных

### Приводим названия столбцов к корректному виду

```python
data = data.rename(columns={
    'airports_nearest' : 'nearest_airport_distance',
    'cityCenters_nearest' : 'city_center_distance',
    'parks_around3000' : 'parks_around_3km',
    'parks_nearest' : 'nearest_park_distance',
    'ponds_around3000' : 'ponds_around_3km',
    'ponds_nearest' : 'nearest_pond_distance'
})
list(data.columns)  # Проверяем новые названия
```

### Обработка пропусков

```python
data.isna().sum()  # Выводим количетсво пропусков
```

```python
data['balcony'] = data['balcony'].fillna(0)
```

Пропуски в столбце `balcony` заменяем на 0, так как, скорее всего, их просто нет.

---

```python
data['is_apartment'] = data['is_apartment'].fillna('False')  
```

Предполагаем, что пропуски в столбце `is_apartment` подразумевают отрицательный ответ и заменяем на False.

---

```python
data['parks_around_3km'] = data['parks_around_3km'].fillna(0)
data['ponds_around_3km'] = data['ponds_around_3km'].fillna(0)
```

Заменим пропуски в `parks_around_3km` и `ponds_around_3km` на 0, так как их просто нет рядом, скорее всего.

---

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Заменим пропуски в `ceiling_height` на медианное значение. Оно в пределах стандартной высоты согласно СНиП.

---

```python
data.groupby('rooms')['living_area'].std()  # Находим стандартное отклонение жилплощади 
                                            # сгруппированное по количеству комнат.
```

Пропуски в `living_area` заменить на медианное значение по количеству комнат не лучшая идея, поскольку стандартное отклонение достаточно велико. Это может исказить данные, относительно зависимости цены.

---

```python
len(data.query('kitchen_area.isna() and (studio == True or open_plan == True or is_apartment == True)'))
```

Количество пропусков по площади кухни в  квартирах-студиях, с открытой планировкой или апартаментах. Оставляем.

---

- Заменили пропуски в столбцах:   
    - `balcony` - заменили на 0, так как, скорее всего, балконов там просто нет.
    - `is_apartment` - очень много пропусков, предполагаем, что это отрицание. Меняем на False. 
    - `parks_around_3km` и `ponds_around_3km` - заменили на 0. Вероятнее всего парков и прудов в округе нет.

### Обработка аномальных значений

```python
data.describe().round(2)
```

`last_price`, `rooms`, `ceiling_height`, `living_area`, `kitchen_area`, `nearest_airport_distance`.

---

#### Last_price

Минимальное значение 12190, максимальное 763млн. 50% наблюдений в диапозоне от 3.4млн до 6.8 млн.

```python
data.hist('last_price', bins=30, range=(0, 1e+06));      # Диапозон от 0 до 1млн
data.hist('last_price', bins=30, range=(1e+08, 8e+08));  # Диапозон от 100млн до 800млн
```

```python
data = data.drop(data[data['last_price'] < 1e+05].index)
data = data.drop(data[data['last_price'] > 1e+08].index)
```

Удаляем квартиры с ценой до 100т и свыше 100млн. От 30 до 100млн будем рассматривать как сегмент элитной недвижимости.

---

#### Rooms

```python
data.query('rooms == 0')['rooms'].count()  # Количество нулевых значений
```

```python
data.query('rooms == 0 and (studio == True or open_plan == True)')
```

Нулевые значения есть либо в квартирах-студиях, либо с открытой планировкой. Оставляем как признак принадлежности к тому или иному виду квартир.

---

```python
data['rooms'].value_counts()
```

```python
data = data.drop(data[data['rooms'] > 9].index)
```

Удаляем квартиры с количеством комнат более 9.

---

#### Ceiling_height

Согласно СНиП минимально допустимая высота потолков в многоквартирных домах 2.5 метра. Стандарт 2.5 - 3.2 метра. Бывают двухуровневые квартиры с высотой 6 метров.

```python
data.query('ceiling_height > 6')
```

```python
data.loc[data['ceiling_height'] > 6, 'ceiling_height'] /= 10
data = data.drop(data[data['ceiling_height'] > 6].index)
data = data.drop(data[data['ceiling_height'] < 2.5].index)
```

Квартиры с высотой потолков до 2.5 метров и более 6 метров удалаяем, так как они не проходят по СНиП. **Кроме тех**, которые в диапозоне от 25 до 60 метров. Их принимаем за опечатку и исправляем, поделив на 10.

---

#### Living_area

Наблюдаются подозрительно маленькие значения.

```python
data.query('living_area < 10')['living_area'].count()  # Количетво квартир с жилой площадью менее 10м
```

```python
data.hist('living_area', range=(0, 20), bins=20);
```

```python
data = data.drop(data[data['living_area'] < 10].index)
```

Удаляем квартиры с жилой площадью менее 10 кв.м.

```python
data = data.drop(data[data['kitchen_area'] < 3].index)
data = data.drop(data[data['kitchen_area'] > 60].index)
```

Удаляем квартиры с площадью кухни меньше 3 и больше 60 кв.м.

    - `kitchen_area` - удалили квартиры с площадью кухни до 3м и свыше 60м.
    - `nearest_airport_distance` - заменили единственное нулевое значение на NaN.

## Создаем новые столбцы

Для удобства нахождения взаимосвязей

### Цена за квадратный метр

```python
data['square_meter_price'] = data['last_price'] / data['total_area']
```

### Дата публикации объявления

Создадим 3 новых столбца. С днем недели, месяцем и годом публикации. И сразу присвоим тип uint8.

```python
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday.astype('uint8')  # День недели
data['exposition_month'] = data['first_day_exposition'].dt.month.astype('uint8')      # Месяц
data['exposition_year'] = data['first_day_exposition'].dt.year.astype('uint16')       # Год
```

### Тип этажа квартиры

Создадим столбец с тремя типами этажности квартир - `первый`, `последний` и `другой`.

```python
floor_compare = []  # Пустой список
for i in range(len(data['floors_total'])):  # Создали цикл с условиями и добавлением категорий в список.
    if data['floor'][i] == 1:
        floor_compare.append('первый')
    elif data['floor'][i] == data['floors_total'][i]:
        floor_compare.append('последний')
    else:
        floor_compare.append('другой')
        
floor_compare = pd.Series(floor_compare)  # Превратили список в Series
```

```python
data['floor_category'] = floor_compare  # Создали новый столбец с категориями
```

```python
data['floor_category'].value_counts()
```

### Расстояние до центра города в километрах

```python
def m_to_km(m):  # Создаем функцию для округления
    try:
        m /= 1000
        return round(m)
    except:
        pass  # На случай NaN
```

```python
data['city_center_distance_km'] = data['city_center_distance'].apply(m_to_km)  # Применяем функцию для создания столбца
```
