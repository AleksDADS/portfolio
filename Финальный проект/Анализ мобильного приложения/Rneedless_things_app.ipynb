### Сессии

**Создадим столбец session_id с уникальным id сессии по каждому пользователю. За границу сессии примем 30 минут**

```python
# сортируем по id и времени для корректности расчёта между соседними событиями
df = df.sort_values(['user_id', 'event_time']).reset_index(drop=True)

# вычисляем разницу во времени между событиями
diff_time = df.groupby('user_id')['event_time'].diff()

# задаём границы между сессиями в 30 минут
new_session = (diff_time.isnull()) | (diff_time > '30 Minutes')

# создаём уникальный id сессии по каждому пользователю
df['session_id'] = df.loc[new_session, ['user_id', 'event_time']] \
    .groupby('user_id').rank(method='first').astype(int)
df['session_id'] = df['session_id'].fillna(method='ffill').astype(int)

df
```

**Создадим датафрейм сессий с временем начала, конца, источником, количеством событий и продолжительности сессии**

```python
# датафрейм
sessions = df.groupby(['user_id', 'session_id']).agg({'event_time':['first', 'last'],
                                                      'source':'first',
                                                      'event_name':'count'}).reset_index()

# переименуем колонки
sessions.columns = ['user_id', 'session_id', 'session_start', 'session_end', 'source', 'events_count']

# добавим продолжительность в секундах
sessions['length'] = (sessions.session_end - sessions.session_start).dt.total_seconds()

sessions
```

### Retention Rate

**Посмотрим на коэффициент удержания для пользователей**

```python
# составим профили пользователей
profiles = sessions.groupby('user_id').agg({'session_start':'first',
                                            'source':'first',
                                            'session_id':'max', 
                                            'length':'sum', 
                                            'events_count':'sum'}).reset_index()

# добавим дату первого посещения
profiles['dt'] = profiles.session_start.dt.date.astype('datetime64')

# переименуем столбцы
profiles.columns = ['user_id', 'first_ts', 'source', 'session_count', 'total_length', 'total_events', 'dt']
```

```python
# добавим маркер show_contacts
users_conversion_list = df.query('event_name == "show_contacts"').user_id.unique()
profiles['show_contacts'] = pd.Series([])
for i in range(len(profiles)):
    if profiles.user_id[i] in users_conversion_list:
        profiles.show_contacts[i] = True
    else:
        profiles.show_contacts[i] = False
```

```python
# создадим функцию для расчёта RR
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # добавляем столбец show_contacts в передаваемый dimensions список
    dimensions = ['show_contacts'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'session_start']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_ts']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # возвращаем таблицу и сырые данные
    return result_raw, result_grouped
```

```python
# рассчитываем RR
result_raw, result_grouped = get_retention(
    profiles, sessions, datetime(2019, 11, 4), 14)

# строим график
plt.figure(figsize=(15, 6)) 
sns.heatmap(result_grouped.drop(columns=['cohort_size', 0]),
            annot=True, fmt='.2%', cbar=False)
plt.title('Коэффициент удержания в разрезе по просмотру контактов')
plt.xlabel('Лайфтайм')
plt.show()
```

## Проверка гипотез

### Гипотеза №1

**Гипотеза H0:** пользователи, из источника "yandex" и "google" демонстрируют **одинаковую** конверсию в просмотр контактов  
**Гипотеза H1:** пользователи, из источника "yandex" и "google" демонстрируют **разную** конверсию в просмотр контактов 

```python
# создадим функцию для проверки Z-Тестом
def stat(successes1, successes2, trials1, trials2):
    alpha = 0.05
    p1 = successes1/trials1
    p2 = successes2/trials2
    p_combined = (successes1 + successes2) / (trials1 + trials2)
    difference = p1 - p2 
    z_value = difference / (p_combined * (1 - p_combined) * (1/trials1 + 1/trials2)) ** 0.5
    distr = st.norm(0, 1)  
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('p_value -', p_value)
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
```

```python
# проверяем
stat(df.query('user_id in @users_conversion_list and source == "yandex"').user_id.nunique(),
     df.query('user_id in @users_conversion_list and source == "google"').user_id.nunique(),
     df.query('source == "yandex"').user_id.nunique(),
     df.query('source == "google"').user_id.nunique())
```

- Разница конверсии в просмотр контактов между пользователями, которые пришли из источника "yandex" и "google" статистически не значима
